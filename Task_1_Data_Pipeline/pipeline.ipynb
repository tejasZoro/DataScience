```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
import seaborn as sns
```

```python
# Step 2: Load the Titanic dataset and save raw data
df = sns.load_dataset('titanic')  # using seaborn's built-in Titanic dataset
df.to_csv("data/raw_titanic.csv", index=False)  # Save original dataset
print(" Raw data saved to data/raw_titanic.csv")
df.head()
```

```python
# Step 3: Check missing values
print("Missing values before handling:\n")
print(df.isnull().sum())
```

```python
# Step 4: Handle missing values

# Numeric columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
imputer_num = SimpleImputer(strategy='mean')
df[num_cols] = imputer_num.fit_transform(df[num_cols])

# Categorical columns
cat_cols = df.select_dtypes(include='object').columns
imputer_cat = SimpleImputer(strategy='most_frequent')
df[cat_cols] = imputer_cat.fit_transform(df[cat_cols])

print("\n✅ Missing values handled.")
```

```python
# Step 5: Encode categorical columns
encoder = LabelEncoder()
for col in cat_cols:
    df[col] = encoder.fit_transform(df[col])

print(" Categorical columns encoded.")
```

```python
# Step 6: Feature scaling for numeric columns
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
print(" Numeric features scaled.")
```

```python
# Step 7: Save cleaned data
df.to_csv("data/clean_titanic.csv", index=False)
print(" Cleaned data saved to data/clean_titanic.csv")

# Optional: Show cleaned data
df.head()
```
